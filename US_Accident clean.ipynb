```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['End_Time'] = pd.to_datetime(df['End_Time'])
# df['Start_Time'] = df['Start_Time'].astype('datetime')
# df['End_Time'] = df['End_Time'].astype('datetime')
```

```python
df_city = df.groupby('City', as_index = False).agg(Accident = ('ID', 'count'))
df_city = df_city.sort_values(['Accident'], ascending = False)
df_city.reset_index(inplace = True, drop = True)
df_city.head(20)
```

```python
df_county = df.groupby('County', as_index = False).agg(Accident = ('ID', 'count'))
df_county = df_county.sort_values(['Accident'], ascending = False)
df_county.reset_index(inplace = True, drop = True)
df_county.head(20)
```

```python
df_state = df.groupby('State', as_index = False).agg(Accident = ('ID', 'count'))
df_state = df_state.sort_values(['Accident'], ascending = False)
df_state.reset_index(inplace = True, drop = True)
df_state.head(20)
```

```python
df_twilight = df.groupby('Civil_Twilight', as_index = False).agg(Accident = ('ID', 'count'))
df_twilight = df_twilight.sort_values(['Accident'], ascending = False)
df_twilight.reset_index(inplace = True, drop = True)
df_twilight.head(20)
```

```python
# df['Start_Hour'] = df['Start_Time'].dt.hour
# df['End_Hour'] = df['End_Time'].dt.hour
df['Start_Hour'] = pd.to_datetime(df['Start_Time']).dt.hour
df['End_Hour'] = pd.to_datetime(df['End_Time']).dt.hour
df['Start_Year'] = pd.to_datetime(df['Start_Time']).dt.year
```

```python
df['Duration'] = df['End_Time'] - df['Start_Time']
```

```python
df_duration = df.groupby('Duration', as_index = False).agg(Accident = ('ID', 'count'))
df_duration = df_duration.sort_values(['Accident'], ascending = False)
df_duration.reset_index(inplace = True, drop = True)
df_duration.head(20)
```

```python
df_severity = df.groupby('Severity', as_index = False).agg(Accident = ('ID', 'count'))
df_severity = df_severity.sort_values(['Accident'], ascending = False)
df_severity.reset_index(inplace = True, drop = True)
df_severity.head(20)
```

```python
df_severity = df.groupby(['Severity','Duration'], as_index = False).agg(Accident = ('ID', 'count'))
df_severity = df_severity.sort_values(['Accident'], ascending = False)
df_severity.reset_index(inplace = True, drop = True)
df_severity.head(20)
```

```python
df_weather = df.groupby('Weather_Condition', as_index = False).agg(Accident = ('ID', 'count'))
df_weather = df_weather.sort_values(['Accident'], ascending = False)
df_weather.reset_index(inplace = True, drop = True)
df_weather.head(20)
```

```python
df_year = df.groupby('Start_Year', as_index = False).agg(Accident = ('ID', 'count'))
df_year = df_year.sort_values(['Accident'], ascending = False)
df_year.reset_index(inplace = True, drop = True)
df_year.head(20)
```

```python
df_start = df.groupby('Start_Hour', as_index = False).agg(Accident = ('ID', 'count'))
df_start = df_start.sort_values(['Start_Hour'], ascending = True)
df_start.reset_index(inplace = True, drop = True)
df_start.head(24)
```

```python
df_start = df.groupby(['Start_Hour','Start_Year'], as_index = False).agg(Accident = ('ID', 'count'))
df_start = df_start.sort_values(['Start_Year'], ascending = True)
df_start.reset_index(inplace = True, drop = True)
df_start.head(20)
```
